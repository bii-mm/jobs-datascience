```python
import pandas as pd
import seaborn as sns
```

```python
safra_anterior = pd.read_csv('Safra_2018-2019.csv')
safra_atual = pd.read_csv('Safra_2020.csv')
```

```python
sns.barplot(x='Categoria_Pesticida', y='Estimativa_de_Insetos', data=safra_anterior)
#é perceptível a existência de uma certa diminuição no número de insetos com o uso frequente de pesticidas
```

```python
sns.barplot(x='dano_na_plantacao', y='Estimativa_de_Insetos', data=safra_anterior)
#Entretanto, no gráfico abaixo percebe-se que nas safras de 2018 e 2019 houveram danos significativos causados pelos 
#pesticidas e que, mesmo com o uso do mesmo, ainda há uma alta estimativa de insetos
```

```python
sns.boxplot(x='dano_na_plantacao', y='Semanas_Utilizando', data=safra_anterior)
#e mesmo durante o uso dos pesticidas, o dano na plantação causado por outros motivos ainda é elevado
```

```python
#a analogia acima fica mais nítida no gráfico abaixo, que mostra que ao decorrer das semanas o uso frequente dos pesticidas
#não diminuiu o dano causado por outros motivos
sns.boxplot(x='dano_na_plantacao', y='Semanas_Utilizando', data=safra_anterior, hue='Categoria_Pesticida')
```

```python
# Traçando um paralelo com a safra de 2020:
sns.kdeplot(safra_anterior['Semanas_Utilizando'])
```

```python
sns.kdeplot(safra_atual['Semanas_Utilizando'])
#a frequência no uso de pesticidas em ambos os dados apresentam um pico na mesma época, o que indica uma provável semelhança
#no resultado dos danos na plantação
```

```python
#Conclusão: 
# É previsto nesse insight que os resultados da safra de 2020 sejam semelhantes aos dos anos anteriores, uma vez que 
#é apresentado uma tendência de uso de pesticidas parecida. Consequentemente, a colheita permanecerá com danos causados
#tanto pelo uso frequente de pesticidas quanto por outros motivos, que também permaneceram persistentes.
```
